```python
def GetNNFeature(input_image):
    input_image = transform.resize(input_image,(640,640,3))
    content_image = reshape_and_normalize_image(input_image)
    sess.run(model['input'].assign(content_image))
    out = model['conv5_4']
    NNFeature = sess.run(out)
    return NNFeature.flatten()
```

```python
imageFeatures = []
for fileName in tqdm.tqdm(allList):
    im = io.imread(fileName)
    F = GetNNFeature(im)
    imageFeatures.append(F)
    
    
```
